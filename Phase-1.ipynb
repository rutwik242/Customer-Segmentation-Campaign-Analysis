```python
# pd.cut() with bin_edges and bin_labels explained:

# bin_edges: Define the actual numeric boundaries where the continuous data (ages) will be split.
# Example: bin_edges = [17, 25, 48, 65]
# - Ages from just above 17 to 25 fall into the first bin.
# - Ages from just above 25 to 48 fall into the second bin.
# - Ages from just above 48 to 65 fall into the third bin.
# These edges control how the data is cut into intervals.

# bin_labels: Define human-readable category names for each bin.
# Example: bin_labels = ['18-25', '26-48', '49-65']
# - Instead of showing intervals like (17, 25], pandas will assign a friendly label like '18-25'.

# Why both are needed:
# - bin_edges specify where to cut the data.
# - bin_labels specify what to name each group after cutting.

# Without bin_labels, pandas will assign default interval ranges like (17.0, 25.0].
# Using bin_labels makes the categories easier to understand in analysis and reports.

# Example:
# df_cust['age group'] = pd.cut(df_cust['age'], bins=bin_edges, labels=bin_labels)
# This creates a new column 'age group' containing labels like '18-25', '26-48', or '49-65'
# based on each customer's age.
```

```python
# Define bin ranges
bin_ranges = [300, 450, 500, 550, 600, 650, 700, 750, 800]

# Create labels for the bins
bin_labels = [f'{start}-{end-1}' for start, end in zip(bin_ranges, bin_ranges[1:])]

# Use pd.cut to assign data to bins
df_cs_clean_1['credit_score_range'] = pd.cut(df_cs_clean_1['credit_score'], bins=bin_ranges, labels=bin_labels, include_lowest=True, right=False)
```

```python
# Define bin ranges to categorize credit scores.
# bin_ranges specify the numeric edges where the data will be split.
# Example: 300-449, 450-499, 500-549, ..., 750-799
#bin_ranges = [300, 450, 500, 550, 600, 650, 700, 750, 800]

# Create labels for each bin to make the categories more understandable.
# bin_labels will look like: '300-449', '450-499', ..., '750-799'
# zip() pairs the start and end of each bin, and f'{start}-{end-1}' creates the label.
#bin_labels = [f'{start}-{end-1}' for start, end in zip(bin_ranges, bin_ranges[1:])]

# Use pd.cut() to assign each customer's credit score to its respective range (bin).
# - df_cs_clean_1['credit_score']: Continuous data to categorize.
# - bins=bin_ranges: Numeric boundaries for the bins.
# - labels=bin_labels: Friendly labels assigned to each bin.
# - include_lowest=True: Ensures that the lowest value (300) is included in the first bin.
# - right=False: Bins are left-inclusive and right-exclusive, i.e., [start, end)
#    Example: 300 is included in 300-449 bin, but 450 is included in 450-499 bin.

# The result is stored in a new column 'credit_score_range',
# where each customer is tagged with their credit score category.


```

```python
# Create human-readable labels for the bins using list comprehension:
# - zip(bin_ranges, bin_ranges[1:]) pairs each start and end of a bin.
#   Example: (300, 450), (450, 500), (500, 550), etc.
# - For each bin, a label is created as 'start-end-1'.
#   Example: for (300, 450), label is '300-449' (because right=False excludes the end).
# - These labels make the bins easy to understand when assigned to the data.

# Example:
# bin_ranges = [300, 450, 500]
# zip gives (300, 450) → label '300-449'
# Then (450, 500) → label '450-499'
# zip(bin_ranges, bin_ranges[1:]) pairs each start and end value for creating bins.
# Example:
# bin_ranges = [300, 450, 500, 550, 600]
# bin_ranges[1:] = [450, 500, 550, 600]
# zip(bin_ranges, bin_ranges[1:]) will produce:
# (300, 450)
# (450, 500)
# (500, 550)
# (550, 600)
# These (start, end) pairs are used to define each bin's range.
```
